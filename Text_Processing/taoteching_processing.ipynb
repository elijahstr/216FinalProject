```python
# We'll also define a method so that we can see the beginning, middle, and end of the text at any given point
def text_peek(text, n=2000, full=True, begin=False, middle=False, end=False):
    if full:
        begin, middle, end = True, True, True
    
    if begin:
        print(text[:n])
        if middle or end:
            print("\n\n------------------------------\n\n")
    
    if middle:
        l = int((len(text)-n)/2)
        print(text[l:l+n])
        if end:
            print("\n\n------------------------------\n\n")
    
    if end:
        print(text[-n:])

text_peek(text)
```

```python
stop_words = set(stopwords.words('english'))
stemmer = SnowballStemmer("english")

def cleanData(text):
    final = []
    for item in text.split():
        x = item.lower()
        x2 = re.sub("[^a-zA-Z]+", "", x)
        tokens = word_tokenize(x2)
        wordlst = [stemmer.stem(word) for word in tokens if word not in stop_words and not word.isdigit()]
        final.extend(wordlst)
    return final

print(cleanData(new_text))
```
