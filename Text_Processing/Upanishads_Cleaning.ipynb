```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer

with open("../Raw_Texts/upanishads.txt", encoding='utf8') as f:
    text = f.read()
```

```python
query = re.compile("Isa-Upanishad")
matches = re.finditer(query, text)
#for match in matches:
   #print(match)
    
query2 = re.compile("Here ends this Upanishad")
matches = re.finditer(query2, text)
#for match in matches:
   #print(match)

isa = text[13012:24294]

headers = ["I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX", "X", "XI", "XII", "XIII", "XIV", "XV", "XVI", "XVII", "XVII"]
for i in headers:
    isa = isa.replace(i, "")
```

```python
query = re.compile("Katha-Upanishad")
matches = re.finditer(query, text)
#for match in matches:
    #print(match)

query2 = re.compile("Here ends this Upanishad")
matches2 = re.finditer(query2, text)
#for i in matches2:
    #print(i)

katha = text[27718:82342]

headers = ["Part First", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX", "X", "XI", "XII", "XIII", "XIV", "XV", "XVI", "XVII", "XVII", "XIX", "XX", "XXI", "XXII", "XXIII", "XXIV", "XXV", "XXVI", "XXVII", "XXVIII", "XXIX", "Part Second", "Part Third", "Part Fourth", "Part Fifth", "Part Sixth"]
for i in headers:
    katha = katha.replace(i, "")
```

```python
query = re.compile("KENA-UPANISHAD")
matches = re.finditer(query, text)
#for match in matches:
    #print(match)

query2 = re.compile("Here ends this Upanishad")
matches2 = re.finditer(query2, text)
#for i in matches2:
    #print(i)
    
kena = text[83027: 101011]

headers = ["Part First", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX", "X", "XI", "XII", "Part Second", "Part Third", "Part Fourth"]
for i in headers:
    kena = kena.replace(i, "")
```

```python
full_text = isa + katha + kena 
with open("upanishads.txt", "w", encoding='utf8') as f:
    f.write(full_text)
```

```python
stop_words = set(stopwords.words('english'))
stemmer = SnowballStemmer("english")

def cleanData(text):
    final = []
    for item in text.split():
        x = item.lower()
        x2 = re.sub("[^a-zA-Z]+", "", x)
        tokens = word_tokenize(x2)
        wordlst = [stemmer.stem(word) for word in tokens if word not in stop_words and not word.isdigit()]
        final.extend(wordlst)
    return final

print(cleanData(full_text))
```
